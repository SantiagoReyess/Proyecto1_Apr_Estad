## REGRESIÓN LINEAL

La regresión lineal es un tipo de modelo estadístico que se usa para relacionar una variable dependiente con una o más variables independientes. Es decir, utiliza variables predictoras para modelar el comportamiento de una variable objetivo. Se llama regresión lineal ya que en su forma más simple, es el ajuste de una línea recta a una nube de puntos, y toma la forma de una ecuación de primer grado. 

## REGRESIÓN POLINOMIAL

La regresión polinomial se desarrolla a partir de la regresión simple, y se encarga de relacionar comportamientos no lineales, en los que una recta común no puede ajustarse de manera adecuada. Normalmente, se elevan los datos de las variables predictoras a la potencia deseada, para obtener un modelo de un grado distinto al lineal y de esta manera captar tendencias diferentes.

## INTERACCIÓN DE FACTORES

La interacción de factores se refiere a la sinergia o la manera en que dos o más variables predictoras interactúan entre sí en un modelo de regresión, en este caso. Para esto, se modela incluyendo términos de interacción en la ecuación de la recta, comúnmente las variables predictoras se multiplican entre sí, y luego se estiman con una beta estimada, para así medir cómo interactúan las variables.  
Es importante para el análisis estadístico entender estas relaciones entre variables predictoras para no sesgar o sobreestimar un modeloa partir de mucha información o relaciones como colinearidad entre variables. 

## SIGNIFICANCIA DE FACTORES

La significancia de factores se refiere al análisis estadístico de los estimadores de un modelo. En el caso de la regresión, los estimadores beta son eso, estimadores obtenidos a partir de una muestra de datos para intentar ajustarse lo mejor posible a una poblacón desconocida. Por ende, los estimadores no son exactos, y utilizando pruebas estadísticas se puede aproximar la veracidad de estos estimadores. Normalmente se calcula el p-value de cada estimador, utilizando pruebas "t" o similares, y se busca un nivel de confianza que satisfacer con estos valores de probabilidad.  

## REGULARIZACIÓN (RIDGE, LASSO, ELASTICNET)

La regularización es una técnica utilizada para reducir el sobreajuste en modelos de regresión. Ridge utiliza la norma L2 para penalizar los coeficientes de una regresión, de manera que no se magnifican pero sin convertirlos en 0. Lasso utiliza la norma L1 para penalizarlos similarmente, pero reduce a cero los coeficientes de las variables menos impactantes del modelo. Finalmente, ElasticNet combina ambos tipos de penalización, impidiendo que se magnifiquen los coeficiente, pero permitiendo seleccionar solo las características importantes del modelo.

## MODELO 1 - TASA DE INTERÉS

El primero modelo, tendrá como objetivo predecir la tasa de interés en base al resto de variables, es decir usaremos la tasa de interés como variable de salida.

```python
#Leemos el dataframe y eliminamos las columnas que no nos interesan del modelo
dataframe = pd.read_csv("LuxuryLoanPortfolio.csv")
dropped =['loan_id', 'funded_date', 'duration years', 'interest rate', 'firstname', 'middlename', 'lastname', 'social', 'phone', 'title',
        'ADDRESS 1', 'ADDRESS 2', 'ZIP CODE', 'CITY', 'STATE', 'TOTAL UNITS', 'LAND SQUARE FEET', 
        'GROSS SQUARE FEET', 'TAX CLASS AT PRESENT', 'BUILDING CLASS CATEGORY', 'BUILDING CLASS AT PRESENT',
         'purpose', 'TAX CLASS AT TIME OF SALE', 'property value', 'payments','employment length']

dataframe = dataframe.drop(columns = dropped)
```

Al analizar el dataframe, podemos ver que hay ciertos datos que no son importantes para nutrir el modelo, por lo que eliminamos las variables que no nos interesan utilizar para modelar. 

### REGRESIÓN SIN PENALIZACIÓN

```python
# Seleccionamos las variables numéricas y categóricas, menos Y

numerical_features = ['funded_amount', 'duration months',  '10 yr treasury index date funded', 
             'total past payments', 'loan balance'] 

categorical_features = []
```

```python
#Definimos las transformaciones que usará el modelo

#categorical_transformer = OneHotEncoder(handle_unknown ="ignore")
numerical_transformer = StandardScaler()
```

```python
#Establecemos el proceso del Pipeline

preprocessor = ColumnTransformer(
    transformers = [
        ('num', numerical_transformer, numerical_features)#,
        #('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
#Elegimos el modelo que usaremos

model = LinearRegression()
```

```python
#Definimos el Pipeline

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', model)]
    )
```

```python
#Seleccionamos nuestro X y Y para el modelo
x = dataframe.drop(columns = ['interest rate percent'])
y = dataframe['interest rate percent']*100

#Separamos en train y test para validar el modelo
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.6, random_state=42)

#Entrenamos el modelo con los valores de train
pipeline.fit(X_train, y_train)
```

```python
#Vizualización de resultados del modelo
#feature_names = (numerical_features + list(pipeline.named_steps['preprocessor']
#                                .named_transformers_['cat']
#                                 .get_feature_names_out(categorical_features)))

feature_names = numerical_features # Cuando no hay columnas categóricas
coefficients = pipeline.named_steps['regressor'].coef_
intercept = pipeline.named_steps['regressor'].intercept_

#Métrica de error
r2_score = pipeline.score(X_test, y_test)
r2_score
```

```python
#Calculo de P-values
y_pred = pipeline.predict(X_train)
residuals = y_train - y_pred

X_train_transformed = pipeline.named_steps['preprocessor'].transform(X_train)#.toarray(
n, k = X_train_transformed.shape

variance = np.sum(residuals**2) / (n - k - 1)

XTX_inv = np.linalg.inv(X_train_transformed.T @ X_train_transformed + np.eye(X_train_transformed.shape[1])*.01)  # (X'X)^(-1)
errores_estandar = np.sqrt(np.diagonal(variance * XTX_inv))

t_stats = coefficients / errores_estandar

p_values = [2 * (1 - stats.t.cdf(abs(t_val), df=n - k - 1)) for t_val in t_stats]

# Print results
print("Intercept:", intercept)
print("Coefficients and P-values:")
for name, coef, p_val in zip(feature_names, coefficients, p_values):
    print(f"  {name}: {coef:.4f} (p-value: {p_val:.4f})")
```
